```python
# Cell 2 — Load the data (FIXED)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Column names
columns = ['unit_id', 'cycle', 'setting1', 'setting2', 'setting3'] + \
          [f'sensor{i}' for i in range(1, 22)]

# Use sep='\s+' to handle multiple spaces, and skip bad rows
train_df = pd.read_csv(
    '../data/raw/train_FD001.txt',
    sep=r'\s+',          # handles multiple spaces
    header=None,
    names=columns,
    engine='python'
)

# Drop fully empty columns
train_df = train_df.dropna(axis=1, how='all')

print(train_df.shape)
print(train_df.head())
```

```python
# Cell 3 — Look at the data
print(train_df.shape)
print(train_df.head())
print(train_df.describe())
```

```python
#cell 4
print(f"Number of engines: {train_df['unit_id'].nunique()}")
print(f"Max cycles: {train_df['cycle'].max()}")
```

```python
engine_1 = train_df[train_df['unit_id'] == 1]
plt.figure(figsize=(12, 4))
plt.plot(engine_1['cycle'], engine_1['sensor2'])
plt.title('Sensor 2 readings for Engine 1 over time')
plt.xlabel('Cycle')
plt.ylabel('Sensor Reading')
plt.show()
```
